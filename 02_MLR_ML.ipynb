```python
# import libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# import your data
df = pd.read_csv("unconv_MV_v5.csv")

df.head()
```

```python
X = df[['Por', 'Perm', 'AI']]
y = df['Prod']
```

```python
# create and fit the model
model = LinearRegression().fit(X, y)
model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# predict the model
model.predict([[12.08, 2.92, 2.80]])
```

```python
# How to plot multiple linear regression
# How to test the Efficacy of the model
# How to check the regression score of the model
```

```python
# split data for training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.2, random_state=5)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# linear regression model for training
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(X_train, y_train)
model
```

# Predict the test results
## Use the predict method to predict the results, then pass the independent variables into it and view the results, it will give the array with all values in it.

```python
y_pred = model.predict(X_test)
y_pred_train = model.predict(X_train)
y_pred_train.shape
```

# Evaluate the Model
## We have different metrics to find the accuracy score of the model and here we use r2_score to evaluate our model and find its accuracy.

```python
# import library for r2 value calculation
from sklearn.metrics import r2_score
accuracy = r2_score(y_test, y_pred)*100
accuracy1 = r2_score(y_train, y_pred_train)*100
print("Accuracy of the test model is %.2f", accuracy)
print("Accuracy of the train model is %.2f", accuracy1)
```

# Plot the results
## We will plot the scatter plot between the actual values and predicted values
 

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred);
plt.xlabel("Actual");
plt.ylabel("Predicted");
```

```python
# data preparation
df = pd.read_csv('unconv_MV_v5.csv')

df.head()
```

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from mpl_toolkits.mplot3d import Axes3D
X = df[['Por', 'Brittle']].values.reshape(-1,2)
Y = df['Prod']
```

```python
# Prepare model for data visualization
x1 = X[:, 0]
x2 = X[:, 1]
# x3 = X[:, 2]
y = Y
x1_pred = np.linspace(6, 24, 30)
x2_pred = np.linspace(0, 100, 30)
# x3_pred = np.linspace(0, 10, 30)
xx1_pred, xx2_pred = np.meshgrid(x1_pred, x2_pred)
model_viz = np.array([xx1_pred.flatten(), xx2_pred.flatten()]).T
```

```python
# Training data
ols1 = linear_model.LinearRegression()
model2 = ols1.fit(X, Y)
predicted1 = model2.predict(model_viz)
```

```python
# Evaluate the results/score
r2 = model2.score(X, Y)
print(r2)
```

```python
plt.style.use('default')

fig = plt.figure(figsize=(12, 4))

ax1 = fig.add_subplot(131, projection='3d')
ax2 = fig.add_subplot(132, projection='3d')
ax3 = fig.add_subplot(133, projection='3d')
# ax4 = fig.add_subplot(134, projection= '3d')

axes = [ax1, ax2, ax3]
# axes = [ax1, ax2, ax3, ax4]

for ax in axes:
    ax.plot(x1, x2, y, color='k', zorder=15, linestyle='none', marker='o', alpha=0.5)
    ax.scatter(xx1_pred.flatten(), xx2_pred.flatten(), predicted1, facecolor=(0,0,0,0), s=20, edgecolor='#70b3f0')
    ax.set_xlabel('Porosity (%)', fontsize=12)
    ax.set_ylabel('Brittle', fontsize=12)
    ax.set_zlabel('Gas Prod. (Mcf/day)', fontsize=12)
    ax.locator_params(nbins=4, axis='x')
    ax.locator_params(nbins=5, axis='x')

ax1.text2D(0.2, 0.32, 'sheharyar', fontsize=13, ha='center', va='center',
           transform=ax1.transAxes, color='Red', alpha=0.5)
ax2.text2D(0.3, 0.42, 'sheharyar', fontsize=13, ha='center', va='center',
           transform=ax2.transAxes, color='Green', alpha=0.5)
ax3.text2D(0.85, 0.85, 'sheharyar', fontsize=13, ha='center', va='center',
           transform=ax3.transAxes, color='Black', alpha=0.5)

ax1.view_init(elev=28, azim=120)
ax2.view_init(elev=4, azim=114)
ax3.view_init(elev=60, azim=165)

fig.suptitle('$R^2 = %.2f$' % r2, fontsize=20)

fig.tight_layout()
```
