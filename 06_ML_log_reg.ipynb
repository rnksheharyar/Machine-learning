```python
# import libraries

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import online data
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# input variables ya features(X)
digits.data.shape
# means 1797 pictures size is 64=8x8
X = digits.data
```

```python
# output labels(y)
digits.target.shape
y = digits.target
```

```python
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:10], digits.target[0:15])):
    plt.subplot(1, 15, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title("Tr: %i\n" % label, fontsize = 20)
```

```python
# help(plt)
```

```python
# split the dat
from sklearn.model_selection import train_test_split
X_trian, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print("Train input data: ", X_trian.shape)
print("Train input data: ", X_test.shape)
print("Train output data: ", y_train.shape)
print("Train output data: ", y_test.shape)
```

```python
# model train
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression().fit(X_trian, y_train)
```

```python
# prediction test
log_reg.predict(X_test[0:5])
predictions = log_reg.predict(X_test)
predictions
```

```python
# accuracy test
score = log_reg.score(X_test, y_test)
print("The accuracy score is: ", score)
```
